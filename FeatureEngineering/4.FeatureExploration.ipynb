```python
#原始的数据路径
OriginDataDir = '../Data/OriginData'
CleanDataDir = '../Data/CleanData'
TrainTestDataDir = '../Data/TrainTestData'
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
from sklearn.preprocessing import PolynomialFeatures
%matplotlib inline
```

```python
df_alltrain = pd.read_csv(os.path.join(TrainTestDataDir,'AllTrain.csv'),index_col=0)
df_alltrain_len = df_alltrain.shape[0]
df_alltrain_target = df_alltrain.TARGET
df_alltrain.drop('TARGET',inplace=True,axis=1)
df_test = pd.read_csv(os.path.join(TrainTestDataDir,'test.csv'),index_col=0)
```

```python
df_train = df_alltrain.append(df_test)
df_train.reset_index(drop=True,inplace=True)
```

```python
df_train.EXT_SOURCE_MIN.head()
```

```python
df_train.shape
```

## 1.EXT_SOURCE_1 EXT_SOURCE_2 EXT_SOURCE_3 DAYS_BIRTH很重要，做polynomial

```python
poly_trans = PolynomialFeatures(degree=3)
```

```python
poly_feature = df_train[['EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3','DAYS_BIRTH']]
```

```python
poly_feature = poly_trans.fit_transform(poly_feature)
```

```python
poly_feature.shape
```

```python
poly_feature
```

```python
polyfeaturename = poly_trans.get_feature_names(input_features=['EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3','DAYS_BIRTH'])
```

```python
df_poly = pd.DataFrame(columns=polyfeaturename,data=poly_feature)
```

```python
df_poly.head()
```

```python
df_poly.drop(['1','EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3','DAYS_BIRTH'],axis=1,inplace=True)
```

```python
df_train = pd.concat([df_train,df_poly],axis=1)
```

```python
df_train.shape
```

## 2.把Flag_Document2-21记起来做一个特征

```python
df_train.filter(regex='FLAG_DOCUMENT*').head()
```

```python
df_train['FLAG_DOCUMENT_SUM'] = df_train.filter(regex='FLAG_DOCUMENT*').sum(axis=1)
```

```python
df_train['FLAG_DOCUMENT_SUM'].head()
```

## 3. 做一些相除的组合特征

```python
df_train['CREDIT_INCOME_PERCENT'] = df_train['AMT_CREDIT']/(df_train['AMT_INCOME_TOTAL']+1e-08)
df_train['ANNUITY_INCOME_PERCENT'] = df_train['AMT_ANNUITY']/(df_train['AMT_INCOME_TOTAL']+1e-08)
df_train['CREDIT_TERM'] = df_train['AMT_ANNUITY']/(df_train['AMT_CREDIT']+1e-08)
df_train['DAYS_EMPLOYED_PERCENT'] = df_train['DAYS_EMPLOYED']/(df_train['DAYS_BIRTH']+1e-08)
```

```python
df_train.shape
```

```python
df_train['AMT_INCOME_perChild'] = df_train['AMT_INCOME_TOTAL']/(df_train['CNT_CHILDREN']+1)
df_train['AMT_INCOME_perDay'] = df_train['AMT_INCOME_TOTAL']/(df_train['DAYS_BIRTH']+1e-08)
df_train['DAYS_EMPLOYED_percent'] = df_train['DAYS_EMPLOYED']/(df_train['DAYS_EMPLOYED']+1e-08)
df_train['OWN_CAR_perDay'] = df_train['OWN_CAR_AGE']/(df_train['DAYS_BIRTH']+1e-08)
df_train['GOODS_INCOME'] = df_train['AMT_GOODS_PRICE'] / (df_train['AMT_INCOME_TOTAL']+1e-08)
df_train['DAY_PHONE_DECISION'] = df_train['DAYS_LAST_PHONE_CHANGE'] / (df_train['DAYS_DECISION_mean']+1e-08)
df_train['INCOME_POPULATION_PERCENT'] = df_train['AMT_INCOME_TOTAL'] / (df_train['REGION_POPULATION_RELATIVE']+1e-08)
```

```python
df_train['MEAN_CACULATE'] = df_train['AMT_INCOME_TOTAL'] - \
df_train['AMT_DRAWINGS_CURRENT_mean_mean'] - df_train['AMT_CREDIT_mean']- df_train['AMT_CREDIT_SUM_DEBT_mean']
```

```python
df_train['MEAN_CACULATE_x'] = (df_train['AMT_DRAWINGS_CURRENT_mean_mean'] + df_train['AMT_CREDIT_mean']+ df_train['AMT_CREDIT_SUM_DEBT_mean'])/(df_train['AMT_INCOME_TOTAL']+1e-08) 
```

## 4.网上看到的一些重要的特征

```python
df_train['NEW_CREDIT_ANNUITY_RATIO'] = df_train['AMT_CREDIT_mean']/(df_train['AMT_ANNUITY']+1e-08)
df_train['NEW_CREDIT_GOODS_RATIO'] = df_train['AMT_CREDIT_mean']/(df_train['AMT_GOODS_PRICE']+1e-08)
df_train['NEW_ANNUITY_INCOME_RATIO'] = df_train['AMT_ANNUITY']/(df_train['AMT_INCOME_TOTAL']+1e-08)
df_train['NEW_PHONE_TO_BIRTH'] = df_train['DAYS_LAST_PHONE_CHANGE']/(df_train['DAYS_BIRTH']+1e-08)
```

```python
df_train.head()
```

## 划分训练集和测试集

```python
from sklearn.model_selection import train_test_split
```

```python
df_test = df_train.iloc[df_alltrain_len:,:]
df_test.reset_index(drop=True,inplace=True)
```

```python
df_train = df_train.iloc[:df_alltrain_len,:]
```

```python
df_train = df_train.sample(frac=1)
df_train.reset_index(drop=True,inplace=True)
```

```python
df_train.shape
```

```python
df_test.head()
```

```python
colwithoutTARGET = df_train.columns.tolist()
colwithTARGET = colwithoutTARGET[:]
colwithTARGET.append('TARGET')
```

```python
print(len(colwithoutTARGET))
print(len(colwithTARGET))
```

```python
# 划分训练集和验证集
X_train, X_val, y_train, y_val = train_test_split(df_train[colwithoutTARGET].values
                                                  ,df_alltrain_target, test_size=0.3,random_state=921)
```

```python
# 划分训练集和验证集
df_train_csv = pd.DataFrame(data=np.hstack([X_train,np.atleast_2d(y_train).T]),columns=colwithTARGET)
df_val_csv = pd.DataFrame(data=np.hstack([X_val,np.atleast_2d(y_val).T]),columns=colwithTARGET)
```

```python
df_train_csv.head()
```

```python
df_val_csv.shape
```

```python
#保存数据
df_train_csv.to_csv(os.path.join(TrainTestDataDir,'train.csv'))
df_val_csv.to_csv(os.path.join(TrainTestDataDir,'validation.csv'))
df_test.to_csv(os.path.join(TrainTestDataDir,'test.csv'))
```
