## 第一次处理credit_card_balance.csv

```python
CreCardBalancePath = os.path.join(OriginDataDir,'credit_card_balance.csv')
df_ccba = pd.read_csv(CreCardBalancePath)
```

```python
df_ccba.dtypes
```

```python
df_ccba.isnull().sum().sort_values(ascending=False)/len(df_ccba)
```

```python
df_ccba_tmp = df_ccba.groupby(['SK_ID_CURR','SK_ID_PREV'])
```

```python
numcollist = df_ccba.select_dtypes(include=[np.number]).columns.tolist()
numcollistwithNULL = df_ccba.loc[:,numcollist].isnull().sum().sort_values(
    ascending=False)/len(df_ccba)
numcollistwithNULL = numcollistwithNULL[numcollistwithNULL>0]
```

```python
numcollistwithNULL.index
```

```python
for col in numcollistwithNULL.index:
    print(col)
    df_ccba[col] = df_ccba_tmp[col].apply(lambda x:x.fillna(x.mean()))
```

#### 处理非数值数据NAME_CONTRACT_STATUS


```python
df_ccba.NAME_CONTRACT_STATUS.value_counts()
```

```python
def changestr(s):
    strset = set(['Active','Completed','Signed','Demand'])
    if not s in strset:
        return 'other'
    else:
        return s
```

```python
#df_ccba.NAME_CONTRACT_STATUS = df_ccba.NAME_CONTRACT_STATUS.apply(changestr)
```

```python
df_ccba = pd.concat([df_ccba,pd.get_dummies(
            df_ccba.NAME_CONTRACT_STATUS,prefix='NAME_CONTRACT_STATUS')],axis=1)
df_ccba.drop(['NAME_CONTRACT_STATUS'],axis=1,inplace=True)
```

### 减少样本量，先合并不同的月份,合并不同的年份记得添加观测了多少个月这个特征

```python
df_ccba.dtypes
```

```python
addmeanmaxmin= ['MONTHS_BALANCE','AMT_BALANCE','AMT_CREDIT_LIMIT_ACTUAL','AMT_INST_MIN_REGULARITY','AMT_PAYMENT_CURRENT'
                ,'AMT_PAYMENT_TOTAL_CURRENT',
                'AMT_TOTAL_RECEIVABLE','CNT_DRAWINGS_ATM_CURRENT','CNT_DRAWINGS_CURRENT','CNT_DRAWINGS_OTHER_CURRENT'
                ,'CNT_DRAWINGS_POS_CURRENT','CNT_INSTALMENT_MATURE_CUM','SK_DPD','SK_DPD_DEF','RECEIVABLE_minus','InterestOfAMT',
               'AMT_DRAWINGS_CURRENT_SUM','AMT_DRAWINGS_ATM_CURRENT','AMT_DRAWINGS_CURRENT',
             'AMT_DRAWINGS_OTHER_CURRENT','AMT_DRAWINGS_POS_CURRENT']
addsum = ['isEqualofAMT_PAYMENT','MONTHS_COUNT_credit_card_balance']
addmean = ['NAME_CONTRACT_STATUS_Active','NAME_CONTRACT_STATUS_Completed','NAME_CONTRACT_STATUS_Sent proposal','NAME_CONTRACT_STATUS_Signed',
          'NAME_CONTRACT_STATUS_Demand','NAME_CONTRACT_STATUS_Approved','NAME_CONTRACT_STATUS_Refused']
```

```python
df_ccba['MONTHS_COUNT_credit_card_balance'] = 1
```

```python
df_ccba.MONTHS_BALANCE = abs(df_ccba.MONTHS_BALANCE)
```

```python
df_ccba_tmp = df_ccba.groupby(['SK_ID_CURR','SK_ID_PREV'])
```

```python
all_COUNT = df_ccba_tmp[addsum].agg('sum')
all_COUNT = all_COUNT.rename(columns={i:i+'_count'for i in all_COUNT.columns})
all_COUNT.head()
```

```python
all_MEAN = df_ccba_tmp[addmean].agg('mean')
all_MEAN = all_MEAN.rename(columns={i:i+'_mean'for i in all_MEAN.columns})
all_MEAN.head()
```

```python
all_mean = df_ccba_tmp[addmeanmaxmin].agg('mean')
all_mean = all_mean.rename(columns={i:i+'_mean'for i in all_mean.columns})
all_mean.head()
```

```python
all_max = df_ccba_tmp[addmeanmaxmin].agg('max')
all_max = all_max.rename(columns={i:i+'_max'for i in all_max.columns})
all_max.head()
```

```python
all_min = df_ccba_tmp[addmeanmaxmin].agg('min')
all_min = all_min.rename(columns={i:i+'_min'for i in all_min.columns})
all_min.head()
```

```python
df_ccba = pd.concat([all_COUNT,all_max,all_mean,all_min,all_MEAN],axis=1)
df_ccba.reset_index(inplace=True)
df_ccba['IsMonthEqual'] = (df_ccba.MONTHS_BALANCE_max!=(df_ccba.MONTHS_COUNT_credit_card_balance_count+1)).astype(int)
df_ccba.head()
```
